# Find Longest Special Substring That Occurs Thrice I

## problem

[Find Longest Special Substring That Occurs Thrice I](https://leetcode.cn/problems/find-longest-special-substring-that-occurs-thrice-i/solutions/2585817/mei-ju-xian-xing-fu-za-du-by-tsreaper-978p/)

You are given a string s that consists of lowercase English letters.
A string is called special if it is made up of only a single character. For example, the string "abc" is not special, whereas the strings "ddd", "zz", and "f" are special.

Return the length of the longest special substring of s which occurs at least thrice, or -1 if no special substring occurs at least thrice.

A substring is a contiguous non-empty sequence of characters within a string.

### Example 1:

Input: s = "aaaa"
Output: 2
Explanation: The longest special substring which occurs thrice is "aa": substrings "aaaa", "aaaa", and "aaaa".
It can be shown that the maximum length achievable is 2.

### Example 2:

Input: s = "abcdef"
Output: -1
Explanation: There exists no special substring which occurs at least thrice. Hence return -1.

### Example 3:

Input: s = "abcaba"
Output: 1
Explanation: The longest special substring which occurs thrice is "a": substrings "abcaba", "abcaba", and "abcaba".
It can be shown that the maximum length achievable is 1.

## resolve

```python
class Solution:
    def maximumLength(self, s: str) -> int:
        # Find all substrings in 's' that are made up of only a single character and record their occurrence times.
        count = Counter()
        occur = 0
        for i,x in enumerate(s):
            if i == 0:
                occur += 1
            elif x == s[i-1]:
                occur += 1
            else:
                #record the occurence count of s[i-1]
                for j in range(occur):
                    count[s[i-1] * (j+1)] += (occur - j)
                #reset the value of occur
                occur = 1
        for j in range(occur):
            count[s[-1] * (j+1)] += (occur - j)
        # traverse the count
        ans = -1
        for element, cnt in count.items():
            if cnt >= 3:
                ans = max(ans,len(element))
        return ans
```
## optimization
Upon analyzing the code,we discovery that recording all length strings consisting of the same single word is unnecessary. A more efficient approach is to only record the occurrence count of the longest string.

For the convenience of counting, we group together the occurrences of strings composed of the same character.

```python
class Solution:
    def maximumLength(self, s: str) -> int:
        # Find all substrings in 's' that are made up of only a single character and record their occurrence times.
        count = defaultdict(list)
        occur = 0
        for i,x in enumerate(s):
            occur += 1
            if i == len(s)-1 or x != s[i+1]:
                count[x].append(occur)
                occur = 0
        ans = 0
        for a in count.values():
            a.sort(reverse=True)
            a.extend([0,0])
            ans = max(ans,a[0]-2,min(a[0]-1,a[1]),a[2])
        return ans if ans else -1
```

## problem
[Number of Boomerangs](https://leetcode.cn/problems/number-of-boomerangs/description/?envType=daily-question&envId=2024-01-08)

You are given n points in the plane that are all distinct, where points[i] = [xi, yi]. A boomerang is a tuple of points (i, j, k) such that the distance between i and j equals the distance between i and k (the order of the tuple matters).

Return the number of boomerangs.

 

Example 1:

Input: points = [[0,0],[1,0],[2,0]]
Output: 2
Explanation: The two boomerangs are [[1,0],[0,0],[2,0]] and [[1,0],[2,0],[0,0]].
Example 2:

Input: points = [[1,1],[2,2],[3,3]]
Output: 2
Example 3:

Input: points = [[1,1]]
Output: 0

## resolve
With a point as the pivot, find two other points such that their distance to the pivot is equal.
When you find a pair of such points, increment the answer by 2.

```python
class Solution:
    def numberOfBoomerangs(self, points: List[List[int]]) -> int:
        n = len(points)
        ans = 0
        # selct a point as a pivot 
        for x1,y1 in points:
            cnt = Counter()
            # Traverse the other points and record each point's distance from the pivot.
            for x2,y2 in points:
                d = (x1-x2) ** 2 +(y1-y2) ** 2
                ans += cnt[d] * 2
                cnt[d] += 1
        return ans
```
